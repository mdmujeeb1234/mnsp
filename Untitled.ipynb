```python
#percentile capping method ( winsorization)
newrr=df['Income']
wupper=np.percentile(newrr,99)
wlower=np.percentile(newrr,1)
outliers=newrr[(newrr<wlower)|(newrr>wupper)]
print("these are the ourliers from Income attribute\n",outliers)
```

```python
df['Income'].plot(kind="hist")
```

```python
for i in df['Income']:
    if i<wlower:
        df["Income"]=df["Income"].replace(i,wlower)
    elif i>wupper:
        df["Income"]=df["Income"].replace(i,wupper)
        
```
